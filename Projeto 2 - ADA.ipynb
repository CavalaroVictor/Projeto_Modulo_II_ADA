### Questão 2

```python
# Criando dataframe de apoio
tabela = pd.DataFrame(columns = ["mandante","mandante_Placar"])
tabela["mandante_Placar"] = tabela["mandante_Placar"].astype("int64")

# Agrupando, por linhas, em um dataframe, cada time mandante em jogos e seus números de gols por jogo
for chave, dataframe in df1.groupby(["mandante"]):
    tabela = pd.concat([tabela, dataframe[["mandante","mandante_Placar"]]])
```

```python
tabela
```

```python
# Sumarizando o número total de gols por time em todos os jogos disputados como mandante
gols_por_time_como_mandante =pd.DataFrame(tabela.groupby("mandante")["mandante_Placar"].sum())
gols_por_time_como_mandante.reset_index(inplace=True)

gols_por_time_como_mandante
```

```python
# Criando dataframe de apoio
tabela = pd.DataFrame(columns = ["visitante","visitante_Placar"])
tabela["visitante_Placar"] = tabela["visitante_Placar"].astype("int64")

# Agrupando, por linhas, em um dataframe, cada time visitante em jogos e seus números de gols por jogo
for chave, dataframe in df1.groupby(["visitante"]):
    tabela = pd.concat([tabela, dataframe[["visitante","visitante_Placar"]]])
```

```python
# Sumarizando o número total de gols por time em todos os jogos disputados como visistante
gols_por_time_como_visitante =pd.DataFrame(tabela.groupby("visitante")["visitante_Placar"].sum())
gols_por_time_como_visitante.reset_index(inplace=True)

gols_por_time_como_visitante
```

```python
# Mesclando os dataframes, para computar nas linhas os gols como mandante e visitante de cada time
gols_por_time = gols_por_time_como_mandante.merge(gols_por_time_como_visitante[["visitante","visitante_Placar"]], how="left", left_on="mandante", right_on="visitante")
```

```python
# Dropando coluna redundante
gols_por_time.drop(columns="visitante",inplace=True)
```

```python
# Criando dataframe com total de gols por time
gols_por_time["total_de_gols"] = gols_por_time["mandante_Placar"] + gols_por_time["visitante_Placar"]
gols_por_time.sort_values("total_de_gols", ascending=False)
```

```python
# Abrindo o dataset das estatísticas das partidas
df_estat = pd.read_csv(r"C:\Users\derik.galante\OneDrive - Hypera\Área de Trabalho\Projeto ADA\campeonato-brasileiro-estatisticas-full.csv", encoding="latin1")
```

## Questão 10

```python
# Tratando None e formato de string de porcentagem para inteiro

df_estat["precisao_passes"] = df_estat["precisao_passes"].str.replace("None","0")
df_estat["precisao_passes"] = df_estat["precisao_passes"].str.replace("%","")
df_estat["precisao_passes"] = df_estat["precisao_passes"].astype("Int64")
```

```python
# Descartando os nulos, agrupando por clube e tirando a média de porcentagem de precisão para cada time
time_mais_preciso =pd.DataFrame(df_estat.drop(df_estat[((df_estat["precisao_passes"].isna()) | (df_estat["precisao_passes"] ==0))].index).groupby("clube")["precisao_passes"].mean())
time_mais_preciso.reset_index(inplace=True)
time_mais_preciso.sort_values(by = "precisao_passes", ascending=False)
```

```python
# Time com mais precisão de passes: Flamengo com 84,5%
```

```python
# Tratando None e formato de string de porcentagem para inteiro

df_estat["posse_de_bola"] = df_estat["posse_de_bola"].str.replace("None","0")

df_estat["posse_de_bola"] = df_estat["posse_de_bola"].str.replace("%","")

df_estat["posse_de_bola"] = df_estat["posse_de_bola"].astype("Int64")
```
